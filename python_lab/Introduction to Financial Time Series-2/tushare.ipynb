## Tushare  https://www.tushare.pro/
Tushare是一个免费、开源的python财经数据接口包。
主要实现对股票等金融数据从数据采集、清洗加工到数据存储的过程，能够为金融分析人员提供快速、整洁、和多样的便于分析的数据，为他们在数据获取方面极大地减轻工作量，使他们更加专注于策略和模型的研究与实现上。考虑到Python pandas包在金融量化分析中体现出的优势，Tushare返回的绝大部分的数据格式都是pandas DataFrame类型，非常便于用pandas/NumPy/Matplotlib进行数据分析和可视化。  

### Tushare获取数据示例：

```python
#导入tushare
import tushare as ts
#设置token，需要声明是谁在调用tushare的api，确认调用者的ID，根据你的积分给相应权限，token号就在tushare主页的个人中心里可以查看，复制粘贴过来
ts.set_token('')
# 
```

```python
#初始化pro接口，把它命名为pro以供调用
pro = ts.pro_api()
```

```python
# A股日线行情--daily
df = pro.daily(ts_code='000001.SZ', start_date='20170101', end_date='20211231')#里面三个参数
#它可以实现批量下载，比如同时下载平安和浦发两只ts_code='000001.SZ, 600000.SH'
df #解读
```

```python
# 利用info获取每个column类型
df.info()
# object暂时可以理解成本字类，无法进行逻辑运算，float浮点数（小数）
```

```python
#针对数字型的column，可以利用 describe 获取数值分布情况。（解读）info和describe在做数据描述性分析的时候经常会用。
df.describe()
```

```python
#数据整理
#后面策略回测的时候惯例时间是由小到大的，而这里大到小，需要给它反过来，我们只需要把它的索引倒过来
df_1 = df.reindex(index=df.index[::-1])#再索引，新索引为原来旧索引的倒序
```

```python
#准备今天的回测数据，青岛啤酒
df_test = pro.daily(ts_code='600600.SH', start_date='20170101',end_date='20211231')
df_1test = df_test.reindex(index=df_test.index[::-1])
#有些列不需要，只保留未来需要的
df_1test.rename(columns={'trade_date':'date','vol':'volume'},inplace = True)#习惯上
df_2test = df_1test[['date','open','high','low','close','volume']] 
df_2test
```

```python
#为了方便以后使用，保存到本地
df_2test.to_csv("600600_daily_2021.csv", index=False)#展示
```
